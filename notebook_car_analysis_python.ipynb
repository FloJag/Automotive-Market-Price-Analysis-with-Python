## Project Goal
This analysis examines the distribution of key features (City MPG, Engine HP) and evaluates the relationships that influence the **Manufacturer's Suggested Retail Price (MSRP)** of vehicles in the dataset.


### The Dataset


*  Make: The manufacturer of the car (e.g., Toyota, Ford, BMW).
*  Model: The specific model name of the car.
*  Year: The production year of the car.
*  Engine Fuel Type: Type of fuel the car uses (e.g., gas, diesel, electric, hybrid).
*  Engine HP: The horsepower of the car's engine.
*  Engine Cylinders: The number of cylinders in the car's engine.
*  Transmission Type: The type of transmission in the car (e.g., automatic, manual).
*   Driven_Wheels: The type of drivetrain (e.g., front-wheel drive, rear-wheel drive, all-wheel drive).
*  Number of Doors: Number of doors the car has (e.g., 2, 4).
*  Market Category: The market segment the car belongs to (e.g., SUV, sedan, luxury).
*  Vehicle Size: The size category of the car (e.g., compact, mid-size, large).
*  Vehicle Style: The style of the vehicle (e.g., coupe, convertible, hatchback).
*  Highway MPG: The estimated miles per gallon on the highway.
*  City mpg: The estimated miles per gallon in the city.
*   Popularity: A numerical value indicating how popular the car is in the market.

## 1. Reading and Cleaning the data

### Import and reading the data

```python
import pandas as pd
df = pd.read_csv('data.csv')
df.head()
```

### Checking the data types

```python
df.dtypes
```

### Identify missing values

```python
df.info()
```

Result: No missing values in the dataset.

### Filtering data

Only include cars from the year 1995 and later

```python
filtered_df = df.loc[df['Year'] > 1994]
filtered_df
```

Test the filter:

```python
print("The max value of the Year is now: ",max(filtered_df['Year']))
print("The min value of the Year is now: ",min(filtered_df['Year']))
```

### Standardize Columns


Make the values of all columns with datatype string in lowercase

```python
columns_to_standardize = ['Make', 'Model', 'Transmission Type', 'Market Category', 'Vehicle Size', 'Vehicle Style']

for column in columns_to_standardize:
  filtered_df.loc[:, column] = filtered_df[column].str.lower()

filtered_df.head()
```

Standardize Columns names with "_" between words

```python
filtered_df.columns = filtered_df.columns.str.replace(' ', '_').str.lower()
filtered_df.head()
```

## 2. Feature Engineering

### Creating New Columns

Create the new column called Total MPG (city mpg + highway mpg)

```python
filtered_df.loc[:, "total_mpg"]= filtered_df["highway_mpg"] + filtered_df["city_mpg"].copy()
filtered_df.head()
```

Create the new column called Price per HP (MSRP / Engine HP)

```python
filtered_df.loc[:, "price_per_HP"] = filtered_df["msrp"] / filtered_df["engine_hp"].copy()
filtered_df.head()
```

## 3. Exploratory Data Analysis

### Descriptive Analysis

Calculate summary statistics for the columns engine_hp, msrp, popularity, highway_mpg, city_mpg.

```python
display(filtered_df[['engine_hp', 'msrp', 'popularity', 'highway_mpg', 'city_mpg']].describe())
```

### Group Analysis

Group the data by driven_wheels, vehicle_size & engine_cylinders and calculate the average MSRP and popularity for each group

```python
grouped_df = filtered_df.groupby(['driven_wheels', 'vehicle_size', 'engine_cylinders'])['msrp'].mean().reset_index()
grouped_df
```

### Visualizations

```python
# Import Libraries

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Histogram

data_column = filtered_df['city_mpg'].dropna()

plt.figure(figsize=(8,5))

plt.hist(data_column, bins=10, alpha=0.7, edgecolor='black', color='red')

plt.xlabel('City MPG')
plt.ylabel('Frequency')
plt.title('Distribution of City MPG')

plt.show()
```

This histogram displays the frequency distribution of the vehicle characteristic City Miles Per Gallon (City MPG).

It highlights that the data is heavily right-skewed (positively skewed), with the vast majority of vehicles achieving a City MPG between approximately 7 and 20 miles per gallon.

The chart confirms that vehicles with very high fuel efficiency (e.g., above 60 MPG) are extremely rare in this dataset, indicating they are significant outliers or belong to a small, specialized segment.

```python
# Bar plot

plt.figure(figsize=(8, 6))

colors = ['red', 'blue', 'green']

sns.barplot(y='msrp', hue='vehicle_size', legend=False,	data=filtered_df, palette=colors)

plt.title('Average MSRP by Vehicle Size')
plt.xlabel('Vehicle Size')
plt.ylabel('Average MSRP')
plt.show()
```

```python
# Scatter Plot

plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='engine_hp',
    y='msrp',
    data=filtered_df,
    color='red',
    marker='o',
    s=100
)
plt.title('Relationship of Engine HP and MSRP')
plt.xlabel('Engine HP')
plt.ylabel('MSRP')
plt.ylim(0, 700000) # Zoom into the chart
plt.show()
```

This scatter plot illustrates the **positive, non-linear relationship** between Engine Horsepower (HP) and the Manufacturer's Suggested Retail Price (MSRP).

The data clearly shows that while higher HP tends to correlate with a higher MSRP, there is a significant price spread for vehicles with high horsepower, indicating other factors heavily influence the final cost.

Consequently, the plot highlights that high HP is a common trait of the most expensive vehicles, but is not the sole determinant of a high price, especially given several Cars with low MSRP and high Engine HP.

```python
plt.figure(figsize=(10, 6))

sns.boxplot(
    data=filtered_df,
    x='driven_wheels',
    y='msrp',
    hue='driven_wheels',
    legend=False,
    palette='Set2'
)


plt.title('Box Plot of MSRP by Driven Wheels')
plt.xlabel('Driven Wheels')
plt.ylabel('MSRP')
plt.ylim(0, 700000) # Zoom into the chart
plt.show()
```

This box plot displays the distribution of MSRP (Manufacturer's Suggested Retail Price) across different Driven Wheels types.

It allows viewers to compare the median price, price spread, and identify potential outliers for vehicles with 'rear wheel drive', 'front wheel drive', 'all wheel drive', and 'four wheel drive'.

For example, it highlights which driven wheel types are associated with higher overall prices or greater price variability.

```python
# Multivariate Line Plot

avg_mpg_by_transmission = filtered_df.groupby('transmission_type')[['city_mpg', 'highway_mpg']].mean().reset_index()

plt.figure(figsize=(10, 6))

sns.lineplot(
    x='transmission_type',
    y='city_mpg',
    data=avg_mpg_by_transmission,
    marker='o',
    label='Average City MPG'
)
sns.lineplot(
    x='transmission_type',
    y='highway_mpg',
    data=avg_mpg_by_transmission,
    marker='o',
    label='Average Highway MPG'
)

plt.title('Average City and Highway MPG by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Average MPG')
plt.legend()
plt.show()
```

This line chart displays the average city and highway MPG across different transmission types.

It clearly shows that highway MPG is higher than city MPG in 4 of the 5 Transmission Types. Viewers can also easily identify which transmission types offer the best or worst fuel efficiency.

### Correlation Analysis

Investigate the correlation between the following variables:
- Engine HP
- MSRP
- Popularity
- City MPG
- Highway MPG

```python
correlation_variables = ['engine_hp', 'msrp', 'popularity', 'city_mpg', 'highway_mpg']
correlation_matrix = filtered_df[correlation_variables].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Selected Variables')
plt.show()
```

The heatmap shows the following correlations:

- Engine HP and MSRP (0.65): There is a strong positive correlation, meaning cars with higher engine horsepower tend to have higher MSRPs.

- City MPG and Highway MPG (0.89): There is a very strong positive correlation, indicating that cars with higher city MPG also tend to have higher highway MPG.

- Engine HP and City/Highway MPG (-0.47, -0.44): There is a moderate negative correlation, suggesting that cars with higher engine horsepower tend to have lower city and highway MPG.

- MSRP and City/Highway MPG (-0.17, -0.18): There is a weak negative correlation, implying that more expensive cars tend to have slightly lower city and highway MPG.

- Popularity: This variable shows very weak correlations with all other variables, indicating that it doesn't have a strong linear relationship with engine HP, MSRP, or fuel efficiency.
